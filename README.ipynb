Ο συντελεστής χωρητικότητας είναι ένας δείκτης απόδοσης των ανανεώσιμων πηγών ενέργειας (ΑΠΕ) που εκφράζει την πραγματική παραγωγή ενέργειας τους ως προς την θεωρητική μέγιστη για μία ορισμένη χρονική περίοδο, εάν η μονάδα ΑΠΕ λειτουργούσε όλη αυτή τη περίοδο στην εγκατεστημένη ισχύ της. <br><br> Προφανώς ο συντελεστής αυτός δεν μπορεί ποτέ να είναι ίσος με τη μονάδα. Για τα ΦΒ πάρκα και τα ΑΠ, ο εξωγενής παράγοντας που καθορίζει τη παραγωγή ενέργειας τους για μια συγκεκριμένη περίοδο είναι οι καιρικές συνθήκες. <br><br> Πέραν αυτού όμως, η θεωρητική μέγιστη ισχύς στον παρανομαστή υπολογίζεται με την παραδοχή ότι τόσο τα ΑΠ όσο και τα ΦΒ λειτουργούν 24h για όλες τις ημέρες αδιάκοπα. Κάτι που για τα ΦΒ πάρκα δεν ισχύει αφού τη νύχτα δεν παράγεται ενέργεια. Ωστόσο ο χρόνος λειτουργίας λαμβάνεται υπόψη στον αριθμητή, μέσω της πραγματικής παραγόμενης ενέργειας. <br> Μαθηματικά, ο συντελεστής αυτός υπολογίζεται ως εξής:

$$ Capacity\:Factor\:(CF)=\frac{Produced\:Energy\:[kWh]}{Installed\:Capacity\:[kW]\times Period\:[h]}\:\leftrightarrow Produced\:Energy\:[kWh]=CF\times Installed\:Capacity\:[kW]\times Period\:[h]$$

Επομένως, με βάση τη παραπάνω σχέση, μπορεί θεωρητικά να υπολογιστεί η παραγωγή ενέργειας μίας μονάδας ΑΠΕ, εάν είναι γνωστή η εγκατεστημένη ισχύς αλλά και ο συντελεστής χωρητικότητας για την εκάστοτε μονάδα. <br> <br> Ως περίοδος χρησιμοποιείται η θεωρητική μέγιστη περίοδος που μπορεί να λειτουργήσει η μονάδα, οπότε δεν χρειάζεται να είναι γνωστή η πραγματική περίοδος λειτουργίας, αφού στον υπολογισμό του αρχικού CF λαμβάνεται υπόψη ο πραγματικός χρόνος λειτουργίας μέσω της παραγόμενης ενέργειας και έτσι ο υπολογισμός γίνεται πιο ακριβής. 

Από τον ΔΑΠΕΕΠ, λαμβάνονται τα απαραίτητα δεδομένα για τις ΦΒ και Αιολικές μονάδες παραγωγής ενέργειας στην Ελλάδα (Ηπειροτική χώρα): Μηνιαία παραγόμενη ενέργεια για κάθε μήνα του έτους 2023 και κατάσταση εγκατεστημένης ισχύος για κάθε μήνα του 2023 και χρησιμοποιούνται για τον υπολογισμό των μηνιαίων συντελεστών χωρητικότητας για κάθε μορφή ενέργειας. <br> <br> Οπότε υπολογίζονται συνολικά 12 CF που προέρχονται από την παραγόμενη ενέργεια ΦΒ δια την εγκατεστημένη ισχύς τους επί τη θεωρητική μέγιστη χρονική περίοδο λειτουργίας και άλλα 12 CF που υπολογίζονται από την εγκατεστημένη ισχύς ΑΠ και την αντίστοιχη παραγόμενη ενέργεια τους ανά μήνα.

Κατόπιν, οι συντελεστές αυτοί χρησιμοποιούνται για την εύρεση της θεωρητικής παραγωγής ενέργειας από τα εμπορικά έργα ενεργειακών κοινοτήτων (1585 ΦΒ πάρκα, 2 ΑΠ και 1 σταθμός βιοαερίου ο οποίος δεν λαμβάνεται υπόψη σε αυτή την ανάλυση), μιας και η εγκατεστημένη ισχύς τους είναι γνωστή από το μητρώο αιτήσεων της χαμηλής-μέσης τάσης του ΔΕΔΔΗΕ. <br><br> Να σημειωθεί εδώ ότι το αποτέλεσμα αποτελεί μία εκτίμηση και δεν αντικατοπτρίζει τη πραγματική ενέργεια που προέρχεται από τις ενεργειακές κοινότητες αφού:

1. Ο συντελεστής χωρητικότητας αποτελεί μία μέση τιμή που αφορά όλα τα ΦΒ (και όλα τα ΑΠ για τους συντελεστές CF των αιολικών) στην Ελλάδα. Αυτό σημαίνει ότι οι ιδιαιτερότητες που καθορίζουν την παραγωγή ενέργειας του κάθε έργου ξεχωριστά αγνοούνται. 

2. Δεν λαμβάνεται υπόψη η υποβάθμιση του βαθμού απόδοσης των μονάδων λόγω φθορών.

3. Δεν λαμβάνεται υπόψη η επιρροή που έχει ο χρόνος διακοπής λειτουργίας στην παραγωγή ενέργειας των μονάδων λόγω συντήρησης αλλά και η διακοπή τους για άλλους λόγους.
